```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

* Display the total number of players


## Player Count

```python
# Get unique players
unique_players_df = pd.DataFrame(purchase_data.drop_duplicates(['SN', 'Gender', 'Age']))
unique_player_count = len(unique_players_df['SN'].unique())

player_count_df = pd.DataFrame({'Total Players': [unique_player_count]})
player_count_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# get the unique items count
unique_items = len(purchase_data['Item ID'].unique())

# find the average price of each purchase
average_price = purchase_data['Price'].mean()

# get the total purchases count
total_purchases = purchase_data['Purchase ID'].count()

# get the total revenue
total_revenue = purchase_data['Price'].sum()


# DataFrame to hold the results
purchase_analysis_totals_df = pd.DataFrame([{'Number of Unique Items': unique_items,
                                            'Average Price': average_price,
                                            'Number of Purchases': total_purchases,
                                            'Total Revenue': total_revenue}])

# Format the columns
purchase_analysis_totals_df['Average Price'] = purchase_analysis_totals_df['Average Price'].map('${:,.2f}'.format)
purchase_analysis_totals_df['Total Revenue'] = purchase_analysis_totals_df['Total Revenue'].map('${:,.2f}'.format)

# Display the results
purchase_analysis_totals_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_count = unique_players_df['Gender'].value_counts()
# gender_count
gender_percentages = unique_players_df['Gender'].value_counts(normalize=True) * 100 
# gender_percentages

# create a data frame to hold the information
gender_demo_df = pd.DataFrame({
    'Total Count': gender_count
   ,'Percentage of Players': gender_percentages 
})

# map the columns to be in a more readable manner
gender_demo_df = gender_demo_df.sort_values('Total Count', ascending = False)
gender_demo_df['Percentage of Players'] = gender_demo_df['Percentage of Players'].map('{:.2f}%'.format)
gender_demo_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Get a clean data set
clean_df = purchase_data[["Gender", "Price"]]

# group the data
clean_gender_group = clean_df.groupby(["Gender"])

# Sum the group of prices
gender_purchase_sum = clean_gender_group['Price'].sum()

# create the results data frame
gender_purchasing_df = pd.DataFrame(purchase_data)
gender_purchase_group = gender_purchasing_df.groupby(['Gender'])
purchase_by_gender_df = pd.DataFrame({
                             'Purchase Count':gender_purchase_group['SN'].count()
                            ,'Average Purchase Price':gender_purchase_group['Price'].mean()
                            ,'Total Purchase Value':gender_purchase_sum                            
                            ,'Avg Total Purchase per Person':gender_purchase_sum / gender_demo_df['Total Count']
})

# Format the columns
purchase_by_gender_df['Average Purchase Price'] = purchase_by_gender_df['Average Purchase Price'].map('${:,.2f}'.format)
purchase_by_gender_df['Total Purchase Value'] = purchase_by_gender_df['Total Purchase Value'].map('${:,.2f}'.format)
purchase_by_gender_df['Avg Total Purchase per Person'] = purchase_by_gender_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)

purchase_by_gender_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# print(purchase_data['Age'].min()) # 7
# print(purchase_data['Age'].max()) # 45

# Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

player_ages_df = pd.DataFrame(unique_players_df)

# Categorize the existing players using the age bins.
player_ages_df['Age Group'] = pd.cut(unique_players_df['Age'], bins, labels=group_names)
player_ages_df['Total Count'] = unique_players_df['Age']

age_percent = player_ages_df['Age Group'].value_counts(normalize=True) * 100


# Create the results
player_age_results_df = pd.DataFrame([])
player_age_results_df['Age Group'] = player_ages_df['Age Group'].value_counts() # player_ages_group['Age Group'].count()
player_age_results_df['Percentage of Players'] = age_percent # player_ages_group['Age Group'].count() / unique_player_count * 100

# Set the sorting
player_age_results_df.sort_index(inplace=True)

# Format the percentage column to 2 decimal places
player_age_results_df['Percentage of Players'] = player_age_results_df['Percentage of Players'].map('{:.2f}%'.format)

player_age_results_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Hold the inital results
top_spenders_df = pd.DataFrame([])
top_spenders_df = purchase_data[['SN', 'Price']]

# Group by the players ids
top_spenders_group = top_spenders_df.groupby(['SN'])

# Create a DataFrame to hold the final results
top_spender_results_df = pd.DataFrame({                            
                            'Purchase Count':top_spenders_group['SN'].count(),
                            'Average Purchase Price':top_spenders_group['Price'].mean(),
                            'Total Purchase Value':top_spenders_group['Price'].sum()
})

# Sort the results in descending order based on Total Purchase Value
top_spender_results_df = top_spender_results_df.sort_values(['Total Purchase Value'], ascending = False)

# Format the data field 
top_spender_results_df['Average Purchase Price'] = top_spender_results_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spender_results_df['Total Purchase Value'] = top_spender_results_df['Total Purchase Value'].map('${:,.2f}'.format)

# Display the results
top_spender_results_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DataFrame based on the purchase data
item_df = pd.DataFrame([])
item_df = purchase_data[['Item ID', 'Item Name', 'Price']]

# Group those items on Item ID and Item Name
item_group = item_df.groupby(['Item ID', 'Item Name'])

# Create a DataFrame to hold the final results
popular_items_results_df = pd.DataFrame([])
popular_items_results_df = item_group.sum()

popular_items_results_df['Purchase Count'] = item_group['Item ID'].count()
popular_items_results_df['Item Price'] = popular_items_results_df['Price'] / popular_items_results_df['Purchase Count']
popular_items_results_df['Total Purchase Value'] = item_group.sum()

# Format the columns
popular_items_results_df['Item Price'] = popular_items_results_df['Item Price'].map('${:,.2f}'.format)
popular_items_results_df['Total Purchase Value'] = popular_items_results_df['Total Purchase Value'].map('${:,.2f}'.format)

# Remove the Price column since it is display in Total Purchase Value
del popular_items_results_df['Price']

# Sort the result according to largest number purchased
popular_items_results_df = popular_items_results_df.sort_values(['Purchase Count'], ascending=False)

# Display the top results
popular_items_results_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
# Copy the previous DataFrame
most_profitable_df = pd.DataFrame(popular_items_results_df)

# Need to reset the Total Purchase Value so it can be sorted
most_profitable_df['Total Purchase Value'] = item_group.sum() 

# Sort the values based on Total Purchase Value
most_profitable_df = most_profitable_df.sort_values(['Total Purchase Value'], ascending=False)

# Format the column
most_profitable_df['Total Purchase Value'] = most_profitable_df['Total Purchase Value'].map('${:,.2f}'.format)

# Display the result
most_profitable_df.head()
```
